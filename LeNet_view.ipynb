# LeNet

### Dataset
* Number of classes: 4
* Classes: *glass*, *metal*, *paper*, *plastic*
* Number of train images for each class: 360
* Number of test images for each class: 40

### Net
* Name: LeNet
* Kernel: 3x3
* MaxPool: 2x2
* Activation functions: ReLU (inside - between Conv and MaxPool), Tanh (after MaxPoll)

### Training
* Number of epochs: 30
* Optimizer: Adam
* Learning rate: 0.001
* Batch size: [80, 64, 32]
* Loss calculation: CrossEntropyLoss

3 trainings, 30 epochs each, same learning rate, different batch sizes.

## Końcowe testowanie mam wrażenie, że jest strasznie losowe. Generalnie w mojej opinii LeNet wychodzi lepiej niż, podczas trenowania czasem nawet accuracy w datasetcie testowym było na poziomie 75%
