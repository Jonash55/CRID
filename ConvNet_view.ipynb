# ConvNet

### Dataset
* Number of classes: 4
* Classes: *glass*, *metal*, *paper*, *plastic*
* Number of train images for each class: 360
* Number of test images for each class: 40

### Net
* Name: ConvNet
* Kernel: 5x5
* MaxPool: 2x2 (Stride: 2)
* Activation functions: ReLU (inside - between Conv and MaxPool)
* Stride: 1
* Padding: 2

### Training
* Number of epochs: 30
* Optimizer: Adam
* Learning rate: 0.001
* Batch size: [80, 64, 32]
* Loss calculation: CrossEntropyLoss

3 trainings, 30 epochs each, same learning rate, different batch sizes.
